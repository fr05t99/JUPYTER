# User Defined Function


```python
def tuble_ex():
    """return defined t tuble"""
    t = (1, 2, 3)
    return t
a, b, c = tuble_ex()
print(a, b, c)
```

```python
# Scope
x = 2
def f():
    x = 3
    return x
print(x) #global scope
print(f()) # local scope
```

```python
x = 5
def f():
    y = 2 * x
    return y
print(f())
#First search local scope, global scope and then built in scope
```

```python
import builtins
dir(builtins) # dir -> returns list of the attributes and methods of any object (say functions , modules, strings, lists, dictionaries etc.)
```

```python
# Nested function
def square():
    """return square of value"""
    def add():
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square())
# search local scope, enclosing function, global and built in scope respectively
```

# Default and Flexible Arguments

```python
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
print(f(5, 4, 3))
```

```python
#flexible arguments *args
def f(*args):
    for i in args:
        print(i)
f(1)
f(1, 2, 3, 4)
#flexible arguments that is dictionary
def f(**kwargs):
    """print key and value of dictionary"""
    for key, value in kwargs.items():
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

# LAMBDA Function

```python
#faster way of writing function
square = lambda x : x ** 2   #where x is name of argument
print(square(4))
tot = lambda x, y, z : x + y + z #where x, y, z are name of arguments
print(tot(1, 2, 3))
```

# Anonymous Function

```python
#like lambda function but can take more than one arguments
# map(func, seq) applies a function to all items in a list
number_list = [1, 2, 3]
y = map(lambda x:x**2, number_list)
print(type(y))
print(y)
print(list(y))
```

# Iterators

```python
# iterable -> an object that can return an iterator
# iterable -> an object with associated iter() method. ex - list, strings
# iterator -> produces next value with next method
name = 'ronaldo'
it = iter(name)
print(next(it))
print(*it)
```

```python
# zip
list1 = [1, 2, 3, 4]
list2 = [5, 6, 7, 8]
z = zip(list1, list2)
print(z)
z_list = list(z)
print(z_list)
```

```python
un_zip = zip(*z_list)
un_list1, un_list2 = list(un_zip) # unzip returns tuple
print(un_list1)
print(un_list2)
print(type(un_list2))
```

# List Comprehension

```python
num1 = [1, 2, 3]
num2 = [i + 1 for i in num1]
print(num2)
```

```python
# Conditionals
num1 = [5, 10, 15]
num2 = [i ** 2 if i == 10 else i - 5 if i < 7 else i + 5 for i in num1]
print(num2)
```

### Exploratory Data Analysis
value_counts() -> frequency counts
outliers -> value that is considerably higher or lower from rest of data

Value at 75% is Q3 and value at 25% is Q1
Outlier are smaller than Q1 - 1.5(Q3 - Q1) and bigger than Q3 + 1.5(Q3 - Q1) 

Called as 1.5(IQR) rule Inner Quartile Range Rule


describe() method ->
25% -> first quantile
50% -> median or second quantile
75% -> third quantile

Quantile ?

1, 4, 5, 6, 8, 9, 11, 12, 13, 14, 15, 16, 17

Median is 11



lower quantile -> median between median and smallest number = 6
upper quantile -> median between median and largest number = 14

```python
print(data['Type 1'].value_counts(dropna = False)) # nan values will also be counted b/c dropna = False
```
